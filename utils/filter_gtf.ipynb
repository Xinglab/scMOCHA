```python
gtf_file = "/home/liuc9/data/refdata/mgatk_index/Homo_sapiens.GRCh38.107.gtf"
```

```python
chrName = "/mnt/isilon/xing_lab/liuc9/refdata/mgatk_index/chrName.txt"
```

```python
new_gtf_file = "/home/liuc9/data/refdata/mgatk_index/Homo_sapiens.GRCh38.107.new.gtf"
```

```python
chrs = []

with open(chrName, "r") as f:
    chrs = [line.strip() for line in f.readlines()]
```

```python
print(chrs)
```

```python
outfile = open(new_gtf_file, "w")

with open(gtf_file, "r") as f:
    for line in f:
        line = line.rstrip()
        if line.startswith("#"):
            outfile.write(line + "\n")
            continue
        
        arr = line.split("\t")
        
        if arr[0] in chrs:
            outfile.write(line + "\n")
outfile.close()
        
```
