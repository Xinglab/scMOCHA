```python
import sys
import re
import pysam
```

```python
bamfile = "/home/liuc9/github/scMOCHA/05-Liming/scmocha-celline/cromwell-executions/scMOCHA/c8470e56-2211-4779-8f58-ee9fc7ddc110/call-cell_cluster_annotation/cacheCopy/execution/MT_cluster.TAG_CJ_cluster_0.bam"
```

```python
bam = pysam.AlignmentFile(bamfile, "rb")
```

```python
NHmax = 1
NMmax = 4
```

```python
def filterReadTags(intags):
    """
    Checks for aligner-specific read tags and filters
    """

    for tg in intags:
        if ("NH" == tg[0] and int(tg[1]) > int(NHmax)) or (("NM" == tg[0] or "nM" == tg[0]) and int(tg[1]) > int(NMmax)):
            return False
    return True
```

```python
intag = []
for read in bam:
    if(filterReadTags(read.tags)):
        intag.append(read.tags)
```
